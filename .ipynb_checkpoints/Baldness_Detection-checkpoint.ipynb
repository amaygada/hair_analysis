```python
import cv2
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from collections import Counter
import os

baldness_status = ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'U1', 'U2', 'U3', 'U4', 'U5', 'U6', 'U7', 'O1', 'O2', 'O3', 'O4', 'O5', 'O6', 'O7', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7']
```

```python
image = cv2.imread('./photos/baldness/1.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, (100, 100), interpolation=cv2.INTER_AREA)
plt.imshow(gray, cmap='gray')
plt.title('my picture')
plt.show()
```

```python
def kmeans(gray, clusters):
    x, y = gray.shape
    gray = gray.reshape(x*y, 1)
    kmeans = KMeans(n_clusters = clusters, n_init = 10)
    kmeans.fit(gray)
    cluster_centers = kmeans.cluster_centers_
    cluster_labels = kmeans.labels_
    image = cluster_centers[cluster_labels].reshape(x,y)
    return image
```

```python
image_4_clusters = kmeans(gray, 4)
plt.imshow(image_4_clusters, cmap='gray')
```

```python
image_2_clusters = kmeans(image_4_clusters, 2)
plt.imshow(image_2_clusters, cmap='gray')
```

```python
image_2_clusters[image_2_clusters > 100] = 0
image_2_clusters[image_2_clusters >0] = 255
```

```python
print(np.unique(image_2_clusters))
```

```python
def denoise(img):
    kernel1 = np.ones((2,2), np.uint8)
    img_erosion = cv2.erode(img, kernel1, iterations=1)
    img_final = cv2.dilate(img_erosion, kernel1, iterations=2)
    
    return img_final

image_final = denoise(image_2_clusters)
plt.imshow(image_final, cmap='gray')
```

```python
def mse(imageA, imageB):
    err = np.sum(((imageA - imageB)/255.0) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err

def most_frequent(List):
    freq = max(set(List), key = List.count)
    if List.count(freq) == 1:
        freq = -1
    return freq
```

```python
path = './photos/baldness/norwood_scale_binary/'
err_status = []
for status in baldness_status:
    img = cv2.imread(path + status + '.png')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    err = mse(img, image_final)
    err_status.append(err)

print(err_status)
```

```python
err_sorted = err_status.copy()
err_sorted.sort()
top_3_index = []

status_final = 0

for i in range(0, 3):
    top_3_index.append(err_status.index(err_sorted[i]))
    
top_3 = []
letter = []
number = []
for i in top_3_index:
    top_3.append(baldness_status[i])
    letter.append(baldness_status[i][0])
    number.append(baldness_status[i][1])

status1 = most_frequent(letter)
hold = []
for i in top_3:
    if i[0] == status1:
        hold.append(i)

top_3_final = hold.copy()

status2 = most_frequent(number)
if status2 != -1:
    for i in top_3_final:
        if i[1] == status2:
            status_final = i
            break

if status_final == 0:
    top_3_final.sort()
    status_final = top_3_final[1]
    
    
status_final
```

```python
output = cv2.imread(path + status_final + '.png')
output = cv2.cvtColor(output, cv2.COLOR_BGR2GRAY)

_, axarr = plt.subplots(1,2)

axarr[0].imshow(image)

axarr[1].imshow(output, cmap='gray')
```
